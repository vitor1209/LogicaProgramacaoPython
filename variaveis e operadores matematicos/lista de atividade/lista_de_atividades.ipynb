## ler 4 numeros e calcular o quadrado e depois a sima de todos

image.png

image.png

```python
n1 = float(input("Digite seu 1° numero: "));
n2 = float(input("Digite seu 2° numero: "));
n3 = float(input("Digite seu 3° numero: "));
n4 = float(input("Digite seu 4° numero: "));

q1 = n1 * n1
q2 = n2 * n2
q3 = n3 * n3
q4 = n4 * n4

print(f"o quadrados dos numeros são:{n1} = {q1:.2f}, {n2} = {q2:.2f}, {n3} = {q3:.2f}, {n4} = {q4:.2f}.");

total = q1 + q2 + q3 + q4;

print(f"E a soma de todos é {total}");
```

## hora extra

image.png

image.png

```python
hora = int(input('Quantas horas você trabalhou (apenas as hora em numero inteiro): '));
mim = int(input('Quantos minutos você trabalhou (apenas as minutos em numero inteiro): '));
valHora = float(input('Quanto você recebe por hora: '));

tempo = (mim / 60) + hora;
salario = tempo * valHora;
horaExtraTotal = (salario * 70 / 100) + salario;

print(f'Seu salario este mês é de: R${salario:.2f}, e seu ganho total com o ajuste de 70% é de: R${horaExtraTotal:.2f}');
```

## velocidade media

image.png

image.png

```python
dist = float(input('Qual e a sua distancia percorrida (em km): '));
hora = int(input('Quantas horas levou (apenas as hora em numero inteiro): '));
mim = int(input('Quantos minutos levou (apenas as minutos em numero inteiro): '));

tempo = (mim / 60) + hora;
vm = dist / tempo;

print(f'Sua velocidade media e de: {vm:.2f} km/h')
```

### velocidade media com condicional

```python
dist = float(input('Qual e a sua distancia percorrida (em km): '));
hora = int(input('Quantas horas levou (apenas as hora em numero inteiro): '));
mim = int(input('Quantos minutos levou (apenas as minutos em numero inteiro): '));

tempo = (mim / 60) + hora;
vm = dist / tempo;

if vm >= 100:
  print(f'Sua velocidade media e de: {vm:.2f} km/h, você esta multado');
else:
  print(f'Sua velocidade media e de: {vm:.2f} km/h');
```
